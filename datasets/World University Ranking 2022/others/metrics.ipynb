```python
import numpy as np
import pandas as pd

voter = np.array(['ARWU-samestruct.xlsx','QS-restruct.xlsx','THE-samestruct.xlsx','URAP-samestruct.xlsx','USNEWS-samestruct.xlsx'])

voters_name = ['ARWU-samestruct', 'QS-restruct', 'THE-samestruct', 'URAP-samestruct', 'USNEWS-samestruct', 'CG', 'borda_score', 'mork_h']
voters = []

for i in range(len(voters_name)):
    voter_path = voters_name[i] + '.xlsx'
    current_voter = pd.read_excel(voter_path, sheet_name="Sheet1")
    voters.append(current_voter)

ARWU = pd.read_excel(voter[0], sheet_name="Sheet1")
QS = pd.read_excel(voter[1], sheet_name="Sheet1")
THE = pd.read_excel(voter[2], sheet_name="Sheet1")
URAP = pd.read_excel(voter[3], sheet_name="Sheet1")
USNEWS = pd.read_excel(voter[4], sheet_name="Sheet1")

def cal_na_group_q(DataFrame):
    NA_group=[]
    DataFrame_dict = DataFrame.to_dict('list')
    for i in range(1,200):
        for n,num in enumerate(DataFrame_dict.get('No')):
            if num == i:
                group = DataFrame_dict.get('Group')[n]
                if group == "North American":
                    NA_group.append(201-i)
    return np.sum(NA_group) if len(NA_group) != 0 else 0
def cal_eu_group_q(DataFrame):
    EU_group=[]
    DataFrame_dict = DataFrame.to_dict('list')
    for i in range(1,200):
        for n,num in enumerate(DataFrame_dict.get('No')):
            if num == i:
                group = DataFrame_dict.get('Group')[n]
                if group == "European":
                    EU_group.append(201-i)
    return np.sum(EU_group) if len(EU_group) != 0 else 0
def cal_asia_group_q(DataFrame):
    ASIA_group=[]
    DataFrame_dict = DataFrame.to_dict('list')
    for i in range(1,200):
        for n,num in enumerate(DataFrame_dict.get('No')):
            if num == i:
                group = DataFrame_dict.get('Group')[n]
                if group == "Asia/Oceania":
                    ASIA_group.append(201-i)
    return np.sum(ASIA_group) if len(ASIA_group) != 0 else 0
def cal_other_group_q(DataFrame):
    Other_group=[]
    DataFrame_dict = DataFrame.to_dict('list')
    for i in range(1,200):
        for n,num in enumerate(DataFrame_dict.get('No')):
            if num == i:
                group = DataFrame_dict.get('Group')[n]
                if group == "other":
                    Other_group.append(201-i)
    return np.sum(Other_group) if len(Other_group) != 0 else 0

def cal_impartiality(DataFrame):
    sum_impartiality = 0.0
    NA_impartiality = 0.0
    EU_impartiality = 0.0
    ASIA_impartiality = 0.0
    Other_impartiality = 0.0
    DataFrame_dict = DataFrame.to_dict('list')
    NA_group=[]
    EU_group=[]
    ASIA_group=[]
    Other_group=[]
    group=[]
    for i in range(1,200): #rank
        for n,num in enumerate(DataFrame_dict.get('No')):
            if num == i:
                group = DataFrame_dict.get('Group')[n]
                if group == "North American":
                    NA_group.append(201-i)
                elif group == "European":
                    EU_group.append(201-i)
                elif group == "Asia/Oceania":
                    ASIA_group.append(201-i)
                else:
                    Other_group.append(201-i)
    data_na = [cal_na_group_q(ARWU),cal_na_group_q(QS),cal_na_group_q(THE),cal_na_group_q(URAP),cal_na_group_q(USNEWS)]
    data_eu = [cal_eu_group_q(ARWU),cal_eu_group_q(QS),cal_eu_group_q(THE),cal_eu_group_q(URAP),cal_eu_group_q(USNEWS)]
    data_asia = [cal_asia_group_q(ARWU),cal_asia_group_q(QS),cal_asia_group_q(THE),cal_asia_group_q(URAP),cal_asia_group_q(USNEWS)]
    data_other = [cal_other_group_q(ARWU),cal_other_group_q(QS),cal_other_group_q(THE),cal_other_group_q(URAP),cal_other_group_q(USNEWS)]  
    NA_impartiality = (np.sum(NA_group) - np.mean(data_na))/np.std(data_na)
    EU_impartiality = (np.sum(EU_group) - np.mean(data_eu))/np.std(data_eu)
    ASIA_impartiality = (np.sum(ASIA_group) - np.mean(data_asia))/np.std(data_asia)
    Other_impartiality = (np.sum(Other_group) - np.mean(data_other))/np.std(data_other)
    sum_impartiality = abs(NA_impartiality) + abs(EU_impartiality) + abs(ASIA_impartiality) + abs(Other_impartiality)
    # print("NA_impartiality: ",NA_impartiality, end='')
    # print("   EU_impartiality: ",EU_impartiality, end='')
    # print("   ASIA_impartiality: ",ASIA_impartiality, end='')
    # print("   Other_impartiality:  ",Other_impartiality, end='')
    return print("   Sum_impartiality: ",sum_impartiality)




for i in range(len(voters)):
    if i % 5 == 0:
        print("-----------------------------------------------------------")
    
    print('cal_impartiality:  Method: %s  eval_result: ' % voters_name[i], end='')
    cal_impartiality(voters[i])
```
